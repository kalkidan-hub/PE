#### Highly Divisible Triangular Number

```python
# a function to calculate number of divisors of a number
def count_divisors(n):
    if n < 1:
        return 0
    count = 0
    for i in range(1, int(n**0.5) + 1):
        if n % i == 0:
            count += 1
            if i != n // i:
                count += 1
    return count
```

```python
number, triangular_number = 1, 1
while True:
    number += 1
    triangular_number += number
    if count_divisors(triangular_number) > 500:
        break

triangular_number
```
