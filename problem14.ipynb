```python
def collatz_seq(n):
    # return number of collatz sequence with n starting number
    count = 1
    while n != 1:
        if n % 2:
            n = n * 3 + 1
        else:
            n //= 2
        count += 1

    return count       
    
    
```

```python
collatz_seq(13)
```

```python
max_collatz, max_collatz_num = 1, 1
for i in range(1, 1_000_000 + 1):
    collatz_num = collatz_seq(i)
    if collatz_num > max_collatz:
        max_collatz, max_collatz_num = collatz_num, i
    
```
