### problem 3

```python
#prime factor of a number
```

```python
x = 600851475143
def prime_factors(n):
    i = 2
    factors = []
    while i * i <= n:
        if n % i:
            i += 1
        else:
            n //= i
            factors.append(i)
    if n > 1:
        factors.append(n)
    return factors
```

### problem 4

```python
#Find the largest palindrome made from the product of two 3-digit numbers
```

```python
def is_palindrome(n):
    return str(n) == str(n)[::-1]
```

```python
# testing all the 3 digit numbers
larg_product = 0
for i in range(999, 99, -1):
    for j in range(999, 99, -1):
        product = i * j
        if is_palindrome(product):
            larg_product = max(larg_product, product)
print(larg_product)
```

```python
from math import gcd

# lcm function
def lcm(*numbers):
    result = 1
    for number in numbers:
        result = result * number // gcd(result, number)
    return result

```

```python
lcm(*range(1,11))
```

```python
lcm(*range(1, 21))
```
