```python
# lattice path through n x n grid
from collections import defaultdict
```

```python
# recursive function to calculate lattice path
def path_count(x, y):
    # base case
    if x == 0 or y == 0: 
        return 1

    return path_count(x - 1, y) + path_count(x, y - 1)

memo = defaultdict(int)
def path_count_with_cache(x,y):
    if x == 0 or y == 0: 
        return 1

    left_point = memo[(x - 1, y)] if (x - 1, y) in memo else path_count_with_cache(x - 1, y)
    up_point = memo[(x, y - 1)] if (x, y - 1) in memo else path_count_with_cache(x, y - 1)
    
    memo[(x,y)] = left_point + up_point

    return  memo[(x,y)]
    
```

```python
path_count(2,2)
```

```python
path_count_with_cache(2,2)
```
