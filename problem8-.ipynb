```python
def sum_primes(n):
    m = (n - 1) // 2
    b = [True] * (m + 1)
    s = 2  # Start with 2, the first prime number
    p = 3
    i = 0

    while p * p < n:
        if b[i]:
            s += p
            j = (p * p - 3) // 2
            while j < m:
                b[j] = False
                j += p
        i += 1
        p += 2

    while i < m:
        if b[i]:
            s += p
        i += 1
        p += 2

    return s

# Example usage
result = sum_primes(2000000)
print(result)
```
